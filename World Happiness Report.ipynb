## TASKS

### To analyse the world happiness data and to find the reason behind the happiness of any country.

```python
sns.pairplot(data_2021[['Ladder score', 'Logged GDP per capita', 'Social support', 'Healthy life expectancy',
                'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']])
plt.suptitle('Pairplot of Happiness Factors (2021)')
plt.show()

# Correlation matrix to quantify relationships
correlation_matrix = data_2021[['Ladder score', 'Logged GDP per capita', 'Social support', 'Healthy life expectancy',
                        'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Relationship Between Happiness Factors (Correlation Matrix)')
plt.show()
```

* The variables included in the pairplot are 'Ladder score' (happiness score), 'Logged GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', and 'Perceptions of corruption'.
* Each variable plotted against every other variable.
* The correlation matrix heatmap provides a visual summary of the correlations between happiness factors.

###  How different factors effect the happiness score.

```python
from sklearn.linear_model import LinearRegression

# Select relevant features and target variable
X = data_2021[['Logged GDP per capita', 'Social support', 'Healthy life expectancy',
        'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']]
y = data_2021['Ladder score']

# Fit a linear regression model
model = LinearRegression()
model.fit(X, y)

# Get coefficients
coefficients = pd.Series(model.coef_, index=X.columns)
print("Regression Coefficients:")
print(coefficients)
```

* These are the coefficients effecting the happiness score.

###  Happiest and Unhappiest Countries in 2021,2022,2023.

```python
happiness_years = ['2021', '2022', '2023']

for year in happiness_years:
    # Filter the data for the specified year
    year_data = df[df['year'] == year]
    
    # Check if there are data points for the specified year
    if not year_data.empty:
        # Get the top 10 and bottom 10 countries by happiness score
        happiest = year_data.nlargest(10, 'Ladder score')
        unhappiest = year_data.nsmallest(10, 'Ladder score')
        
        # Create bar plots for the happiest and unhappiest countries
        plt.figure(figsize=(10, 6))
        sns.barplot(x='Ladder score', y='Country name', data=happiest, palette='coolwarm')
        plt.xlabel('Happiness Score')
        plt.title(f'Happiest Countries in {year}')
        plt.show()
    else:
        print(f"No data available for the year {year}")
```
